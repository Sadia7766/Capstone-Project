```python
# ==============================================================
# 1️⃣ Mount Drive & Unzip Dataset
# ==============================================================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

!unzip -q "/content/drive/MyDrive/Zip/Bangla sign dataset.zip" -d "/content/Bangla sign dataset.zip"

# ==============================================================
# 2️⃣ Imports & Config
# ==============================================================

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os, random, shutil
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
from collections import Counter

# --------------------------------------------------------------
# USER CONFIG
# --------------------------------------------------------------
DATA_ROOT = "/content/Bangla sign dataset.zip"
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42
NUM_EPOCHS = 40
MODEL_OUTPUT = "handsign_balanced_model.h5"
AUTO_TUNE = True
TARGET_SIZE = 300   # ✅ Minimum number of images per class after augmentation

tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

print("TensorFlow:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))

# ==============================================================
# 3️⃣ FIX EXTRA NESTED FOLDER (if any)
# ==============================================================

inner_dirs = [os.path.join(DATA_ROOT, d) for d in os.listdir(DATA_ROOT) if os.path.isdir(os.path.join(DATA_ROOT, d))]
if len(inner_dirs) == 1 and any(os.path.isdir(os.path.join(inner_dirs[0], x)) for x in os.listdir(inner_dirs[0])):
    print("📁 Flattening nested folder structure...")
    nested_root = inner_dirs[0]
    for cls in os.listdir(nested_root):
        shutil.move(os.path.join(nested_root, cls), os.path.join(DATA_ROOT, cls))
    shutil.rmtree(nested_root)

# ==============================================================
# 4️⃣ Count Classes and Samples
# ==============================================================

def count_images(root):
    total = 0
    class_counts = {}
    for cls in sorted(os.listdir(root)):
        cls_path = os.path.join(root, cls)
        if os.path.isdir(cls_path):
            num = len([f for f in os.listdir(cls_path) if f.lower().endswith(('jpg','jpeg','png'))])
            class_counts[cls] = num
            total += num
    return total, class_counts

total_imgs, class_counts = count_images(DATA_ROOT)
print(f"\n📊 Found total {total_imgs} images across {len(class_counts)} classes:")
for cls, num in class_counts.items():
    print(f" - {cls}: {num} images")

# ==============================================================
# 5️⃣ Balance Dataset
# ==============================================================

print(f"\n⚙️ Balancing dataset — ensuring at least {TARGET_SIZE} images per class...")

aug_dir = "/content/balanced_dataset"
if os.path.exists(aug_dir):
    shutil.rmtree(aug_dir)
os.makedirs(aug_dir, exist_ok=True)

datagen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.25,
    shear_range=0.25,
    brightness_range=[0.7, 1.3],
    horizontal_flip=True,
    fill_mode="nearest"
)

for cls, count in class_counts.items():
    src_dir = os.path.join(DATA_ROOT, cls)
    dst_dir = os.path.join(aug_dir, cls)
    os.makedirs(dst_dir, exist_ok=True)

    # Copy originals
    for img_file in os.listdir(src_dir):
        if img_file.lower().endswith(('jpg','jpeg','png')):
            shutil.copy(os.path.join(src_dir, img_file), os.path.join(dst_dir, img_file))

    # Augment to reach TARGET_SIZE
    if count < TARGET_SIZE:
        num_needed = TARGET_SIZE - count
        img_list = [os.path.join(src_dir, f) for f in os.listdir(src_dir) if f.lower().endswith(('jpg','jpeg','png'))]
        i = 0
        for img_path in img_list:
            img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
            x = tf.keras.preprocessing.image.img_to_array(img)
            x = np.expand_dims(x, 0)
            for _ in datagen.flow(x, batch_size=1, save_to_dir=dst_dir, save_prefix="aug", save_format="jpg"):
                i += 1
                if i >= num_needed:
                    break
            if i >= num_needed:
                break

balanced_total, balanced_counts = count_images(aug_dir)
print(f"\n✅ After balancing: {balanced_total} total images")
for cls, num in balanced_counts.items():
    print(f" - {cls}: {num} images")

# ==============================================================
# 6️⃣ Train / Validation Split
# ==============================================================

train_ds = tf.keras.utils.image_dataset_from_directory(
    aug_dir,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    aug_dir,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_ds.class_names
num_classes = len(class_names)
print(f"\n✅ Detected classes: {class_names}")

if AUTO_TUNE:
    AUTOTUNE = tf.data.AUTOTUNE
    train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
    val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

# ==============================================================
# 7️⃣ Define Model (MobileNetV2)
# ==============================================================

base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==============================================================
# 8️⃣ Train Model
# ==============================================================

history = model.fit(
    train_ds,
    epochs=NUM_EPOCHS,
    validation_data=val_ds
)

# ==============================================================
# 9️⃣ Fine-tune Backbone
# ==============================================================

print("\n🔧 Starting safe fine-tuning...")

base_model.trainable = True
for layer in base_model.layers:
    if isinstance(layer, tf.keras.layers.BatchNormalization):
        layer.trainable = False

for layer in base_model.layers[:-20]:
    layer.trainable = False

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("best_finetuned_model.h5", save_best_only=True)
]

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-6),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=callbacks
)

# ==============================================================
# 🔟 Evaluate & Save
# ==============================================================

val_loss, val_acc = model.evaluate(val_ds)
print(f"\n✅ Final Validation Accuracy: {val_acc*100:.2f}%")

model.save(MODEL_OUTPUT)
print(f"\n💾 Model saved as {MODEL_OUTPUT}")

# ==============================================================
# 11️⃣ Plot Accuracy/Loss
# ==============================================================

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'] + history_fine.history['accuracy'])
plt.plot(history.history['val_accuracy'] + history_fine.history['val_accuracy'])
plt.title('MobileNetV2 accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['train', 'val'])

plt.subplot(1,2,2)
plt.plot(history.history['loss'] + history_fine.history['loss'])
plt.plot(history.history['val_loss'] + history_fine.history['val_loss'])
plt.title('MobileNetV2 loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['train', 'val'])
plt.show()

# ==============================================================
# 12️⃣ VGG16 Model
# ==============================================================

from tensorflow.keras.applications import VGG16

print("\n🚀 Training VGG16 model...")

base_vgg = VGG16(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
base_vgg.trainable = False

model_vgg = models.Sequential([
    base_vgg,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model_vgg.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_vgg = model_vgg.fit(
    train_ds,
    validation_data=val_ds,
    epochs=NUM_EPOCHS
)

print("\n🔧 Fine-tuning VGG16...")
base_vgg.trainable = True
for layer in base_vgg.layers[:-10]:
    layer.trainable = False

callbacks_vgg = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("best_vgg16_model.h5", save_best_only=True)
]

model_vgg.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-6),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_vgg_fine = model_vgg.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=callbacks_vgg
)

vgg_val_loss, vgg_val_acc = model_vgg.evaluate(val_ds)
print(f"\n✅ Final VGG16 Validation Accuracy: {vgg_val_acc*100:.2f}%")

model_vgg.save("handsign_vgg16_model.h5")
print("💾 VGG16 model saved as handsign_vgg16_model.h5")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

def evaluate_model(model, dataset, model_name):
    print(f"\n==============================")
    print(f"📌 Results for: {model_name}")
    print(f"==============================")

    # Extract true labels
    y_true = []
    y_pred = []

    for images, labels in dataset:
        preds = model.predict(images)
        preds = np.argmax(preds, axis=1)

        y_true.extend(labels.numpy())
        y_pred.extend(preds)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Classification Report
    print(f"\n📄 Precision–Recall–F1 Report ({model_name}):")
    print(classification_report(y_true, y_pred, target_names=class_names))

    return cm

# ==============================================================
# Generate Results For VGG16 Only
# ==============================================================

# Only evaluate VGG16 (the model with highest accuracy)
cm_vgg16 = evaluate_model(model_vgg, val_ds, "VGG16")
```

```python
# ==============================================================
# 16️⃣ Real-time Webcam Prediction in Colab
# ==============================================================

import cv2
import numpy as np
import IPython
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import time

# Function to capture image from webcam
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
        const div = document.createElement('div');
        const capture = document.createElement('button');
        capture.textContent = 'Capture';
        div.appendChild(capture);

        const video = document.createElement('video');
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});

        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();

        // Resize video for better display
        video.style.width = '640px';
        video.style.height = '480px';

        // Wait for Capture button to be clicked
        await new Promise((resolve) => {
            capture.onclick = resolve;
        });

        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;
        canvas.getContext('2d').drawImage(video, 0, 0);
        stream.getVideoTracks()[0].stop();
        div.remove();

        return canvas.toDataURL('image/jpeg', quality);
    }
    ''')

    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])

    with open(filename, 'wb') as f:
        f.write(binary)

    return filename

# Function to make prediction
def predict_from_image(image_path, model, class_names, img_size=(224, 224)):
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=img_size)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Add batch dimension

    # Make prediction
    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions[0])]
    confidence = np.max(predictions[0]) * 100

    return predicted_class, confidence, predictions[0]

# Main prediction loop
def run_webcam_prediction(model, class_names, img_size=(224, 224), num_predictions=10):
    print("📸 Webcam Prediction System")
    print("=" * 40)
    print(f"Model: {model_path}")
    print(f"Classes: {', '.join(class_names)}")
    print(f"Will take {num_predictions} predictions")
    print("=" * 40)

    for i in range(num_predictions):
        print(f"\n📸 Prediction #{i+1}/{num_predictions}")
        print("Click the 'Capture' button when ready...")

        # Take photo
        photo_filename = f'capture_{i}.jpg'
        take_photo(photo_filename)

        # Make prediction
        predicted_class, confidence, all_probs = predict_from_image(
            photo_filename, model, class_names, img_size
        )

        # Display results
        print(f"✅ Predicted: {predicted_class}")
        print(f"📊 Confidence: {confidence:.2f}%")

        # Show top 3 predictions
        top_3_indices = np.argsort(all_probs)[-3:][::-1]
        print("\nTop 3 predictions:")
        for idx in top_3_indices:
            class_name = class_names[idx]
            prob = all_probs[idx] * 100
            print(f"  - {class_name}: {prob:.2f}%")

        # Display the captured image
        img = plt.imread(photo_filename)
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.title(f'Prediction: {predicted_class} ({confidence:.2f}%)')
        plt.axis('off')
        plt.show()

        time.sleep(1)  # Small delay between predictions

# Run the webcam prediction
print("🚀 Starting webcam prediction...")
run_webcam_prediction(best_model, class_names, IMG_SIZE, num_predictions=5)
```
